```python
torch.cuda.is_available()
```

```python
x = torch.rand(5,3)
```

```python
print("X in CPU memory\n", x)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

if torch.cuda.is_available():
  y = torch.ones_like(x, device=device)
  print("y in CPU memory\n", y)
  x = x.to(device)
  print("x in GPU memory\n", x)
  z = x + y
  print("Z in GPU Memory after addition\n", z)
  print(z.to("cpu", torch.double))

```

```python
x = torch.rand(5,3)
print("X in CPU memory\n", x)

device = 'cuda' if torch.cuda.is_available() else 'cpu'

if torch.cuda.is_available():
  y = torch.ones_like(x, device=device)
  print("y in CPU memory\n", y)
  #x = x.to(device)
  #print("x in GPU memory\n", x)
  z = x + y
  print("Z in GPU Memory after addition\n", z)
  print(z.to("cpu", torch.double))
```
