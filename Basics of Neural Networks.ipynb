```python
import torch
```

```python
targets = torch.tesnor([1,  0, 1])
preds = torch.tensor([0.9, 0.8, 0.2])
```

```python
def loss_func(prediction, target):
  return torch.where(target == 1, 1-prediction, prediction).mean()
# The problem with this loss func is that it assumes that predicition is always between 0 and 1. And we have to make sure that it happens.
#Alternate for this is a SIGMOID Func 
```

```python
loss_func(preds, targets)
```

```python
targets = torch.tensor([1,  0, 1])
preds = torch.tensor([0.9, 0.1, 0.8])
loss_func(preds, targets)
```

```python
# With these 2 examples we can observe that loss func returns a lowe number if predictions are more accurate
```

```python
import matplotlib.pyplot as plt
def plot(f, min=-5, max=5, figsize=(6,4)):
  x = torch.linspace(min, max)
  fig, ax = plt.subplots(figsize = figsize)
  ax.plot(x, f(x))

```
