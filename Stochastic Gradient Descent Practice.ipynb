```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def f(x): return x**2
```

```python
def plot_function(f, tx=None, ty= None, min=-2, max=2, figsize=(6,4)):
  x = torch.linspace(min, max)
  fig,ax = plt.subplots(figsize=figsize)
  ax.plot(x, f(x))
  ax.set_xlabel(tx)
  ax.set_ylabel(ty)
```

```python
plot_function(f, "x", "x**2")
```
